```python
def make_genres_list(x):
  gen=[]
  st=' ' # space , to put between genres

  for i in x:
    if i.get('name')=='Science Fiction':
      scifi='Sci-Fi'
      gen.append(scifi)
    else:
      gen.append(i.get('name'))
  if gen==[]:
    return np.NaN
  else:
    return (st.join(gen))
```

```python
data['genres']=data['genres'].map(lambda x:make_genres_list(x))
```

```python
data['genres']
```

```python
def get_actor1(x):
  casts=[]
  for i in x:
    casts.append(i.get('name'))
  if casts==[]:
    return np.NaN
  else:
    return (casts[0])
```

```python
data['actor_1_name']=data['cast'].map(lambda x:get_actor1(x))
```

```python
def get_actor2(x):
  casts=[]
  for i in x:
    casts.append(i.get('name'))
  if casts==[] or len(casts)<=1:
    return np.NaN
  else:
    return (casts[1])
```

```python
data['actor_2_name']=data['cast'].map(lambda x:get_actor2(x))
```

```python
def get_actor3(x):
  casts=[]
  for i in x:
    casts.append(i.get('name'))
  if casts==[] or len(casts)<=2:
    return np.NaN
  else:
    return (casts[2])
```

```python
data['actor_3_name']=data['cast'].map(lambda x:get_actor3(x))
```

```python
def get_directors(x):
  dt=[]
  st=' '
  for i in x:
    if i.get('job')=='Director':
      dt.append(i.get('name'))
    if dt==[]:
      return np.NaN
    else:
      return (st.join(dt))
```

```python
from tmdbv3api import Movie
import numpy as np
import requests
import time

tmdb_movie = Movie()

def get_genre(x):
    try:
        # Add a small delay to avoid hitting API rate limits
        time.sleep(0.25)

        # Handle empty or invalid titles
        if pd.isna(x) or x == '':
            return np.NaN

        # Search for the movie
        result = tmdb_movie.search(x)

        if not result:
            return np.NaN

        # Get the first movie result
        movie_id = result[0].id

        # Make API request for movie details
        response = requests.get(
            f'https://api.themoviedb.org/3/movie/{movie_id}',
            params={'api_key': tmdb.api_key}
        )

        # Check if request was successful
        if response.status_code != 200:
            return np.NaN

        data_json = response.json()

        # Extract genres
        if data_json.get('genres'):
            genres = [genre['name'] for genre in data_json['genres']]
            return ' '.join(genres)

        return np.NaN

    except Exception as e:
        print(f"Error processing title '{x}': {str(e)}")
        return np.NaN

# Apply the function to the dataframe with progress tracking
from tqdm import tqdm
tqdm.pandas()
df_2020['genres'] = df_2020['Title'].progress_map(lambda x: get_genre(str(x)))
```
